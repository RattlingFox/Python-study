##### <font color='red'>Задание</font>

Создать класс, реализующий функционал учета рейтинга студентов **StudentsRaiting**. Класс должен иметь конструктор, который будет принимать список студентов с результатами сданных экзаменов, реализованый в виде словаря с ключем "ФИО студента" и оценками по экзаменам по 4 предметам (Программирование, Дискретная математика, Иностранный язык, Философия), представленными в виде упорядоченного множества (кортежа). Например:

`{"Иванов Петр Николавич":(3,3,5,4), "Петров Иван Константинович":(5,4,5,4)}`

Релизовать методы: 
- добавления студента с оценками по экзаменам
- удаления студента из рейтинга 
- получения списка студентов с рейтингом выше порогового
- метод демонстрации работы методов класса с выводом результатов на печать в виде текстовой таблицы.

Примечание:
Рейтинг составляется по суммарному баллу по всем предметам.

```python
from tabulate import tabulate
import numpy as np

class StudentsRaiting:

    student = dict = {}
        
    def set_info(self):
        print("Введите ФИО")
        name=input()
        i=1
        mark_i=[]
        temp=0
        while i<=4:
            print("Введите оценку № ", i)
            temp=input()
            if temp=='' or temp==None:
                temp=0
            mark_i.append(int(temp))
            i = i + 1
        mark=tuple(mark_i)
        self.student.update({name:mark})
    
    def del_info(self):
        print("Введите ФИО для удаления")
        name=input()
        self.student.pop(name)
    
    def look_info(self):
        print(self.student)
        
    def look_top(self):
        top=0
        print("Введите пороговое значение")
        top=int(input())
        result = {}
        for name,mark in self.student.items():
            if sum(mark)>=top:
                result.update({name:mark})
        print(result)
    def demo(self):
        self.student.update({"First Demo Student":np.random.randint(0, 5, 4),
                     "Second Demo Student":np.random.randint(0, 5, 4),
                     "Third Demo Student":np.random.randint(0, 5, 4),
                     "Fourth Demo Student":np.random.randint(0, 5, 4)})
    def look_table(self):
        from tabulate import tabulate
        print(tabulate(self.student.items(), headers=['NAME', 'MARK'], tablefmt="grid"))
            
s = StudentsRaiting()
pi = 0
while pi < 1:
    print("Введите название операции (Добавить, Удалить, Показать, Рейтинг, Таблица, Демо, Выход):")
    a=input()
    if a=="Добавить":
        s.set_info()
    elif a=="Удалить":
        s.del_info()
    elif a=="Показать":
        s.look_info()
    elif a=="Рейтинг":
        s.look_top()
    elif a=="Таблица":
        s.look_table()
    elif a=="Демо":
        s.demo()
    elif a=="Выход":
        pi = 1
    else:
        print('Неверная команда!')
```
